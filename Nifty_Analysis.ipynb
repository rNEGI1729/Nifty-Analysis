# Task 2
- Develope SMA Crossover Strategy, which is Sell when short term MA crosses above long term MA and buy when short term MA crosses below long term MA.
- For 5 stocks
- short term MA - 50 days
- long term MA - 200 days
- Backtest this strategy on data to built buy/sell signals and see if whether I would've made profit or loss.



```python
# 5 stocks
tickers = ['RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'INFY.NS', 'ICICIBANK.NS']

# Define the start and end dates for the last 12 years
end_date = pd.Timestamp.today()
start_date = end_date - pd.DateOffset(years=12)

# Fetching data for the last 12 years
stock_data1 = yf.download(tickers, start=start_date, end=end_date, group_by='ticker')

# Display the first few rows of the data
stock_data1.head()
```

```python
# Calculating 50_MA (short term MA)and 200_MA (long term MA)

for ticker in tickers:
  stock_data1.loc[:,(ticker,'short_term_MA')] = stock_data1[ticker]['Close'].rolling(window=50).mean()
  stock_data1.loc[:,(ticker,'long_term_MA')] = stock_data1[ticker]['Close'].rolling(window=200).mean()

```

```python

for ticker in tickers:

  # Create a signal: Buy when 50_MA crosses above 200_MA, sell when it crosses below
  stock_data1.loc[:,(ticker,'Signal')] = 0.0
  stock_data1.iloc[100:, stock_data1.columns.get_loc((ticker, 'Signal'))] = np.where(
    stock_data1.loc[:,(ticker,'short_term_MA')][100:] > stock_data1.loc[:,(ticker,'long_term_MA')][100:], 1.0, 0.0)

  # Generate buy/sell signals
  stock_data1.loc[:,(ticker,'Position')] = stock_data1.loc[:,(ticker,'Signal')].diff() # +1 for buying and -1 for selling

  # Plot buy and sell signals
  plt.figure(figsize=(10, 6))
  plt.plot(stock_data1.loc[:,(ticker,'Close')], label=f'{ticker} Closing Price')
  plt.plot(stock_data1.loc[:,(ticker,'short_term_MA')], label='50-day MA')
  plt.plot(stock_data1.loc[:,(ticker,'long_term_MA')], label='200-day MA')

  # Plot Buy signals
  plt.plot(stock_data1[ticker][stock_data1.loc[:,(ticker,'Position')] == 1].index,
         stock_data1.loc[:,(ticker,'Close')][stock_data1.loc[:,(ticker,'Position')] == 1],
         '^', markersize=10, color='g', lw=0, label='Buy Signal')

  # Plot Sell signals
  plt.plot(stock_data1[ticker][stock_data1.loc[:,(ticker,'Position')] == -1].index,
         stock_data1.loc[:,(ticker,'Close')][stock_data1.loc[:,(ticker,'Position')] == -1],
         'v', markersize=10, color='r', lw=0, label='Sell Signal')

  plt.title('Moving Average Crossover Strategy - Buy & Sell Signals')
  plt.legend()
  plt.show()
```

```python
for ticker in tickers:

  # Calculate returns
  stock_data1['Market Return'] = stock_data1[ticker]['Close'].pct_change()
  stock_data1['Strategy Return'] = stock_data1['Market Return'] * stock_data1[ticker]['Signal'].shift(1)  # shifting rows by one as realistically we can pick stock on next day

  stock_data1['Cumulative Market Return'] = (1 + stock_data1['Market Return']).cumprod()
  stock_data1['Cumulative Strategy Return'] = (1 + stock_data1['Strategy Return']).cumprod()

  # Display results
  plt.figure(figsize=(12, 6))
  plt.plot(stock_data1['Cumulative Market Return'], label='Buy and Hold', color='blue')
  plt.plot(stock_data1['Cumulative Strategy Return'], label='SMA Crossover Strategy', color='orange')
  plt.title(f"Cumulative Returns: {ticker}")
  plt.xlabel("Date")
  plt.ylabel("Cumulative Returns")
  plt.legend()
  plt.show()
```

# Task 3

- Calculate Maximum Drawdown for Each Stock over 6-month period.
- Implement Stop-Loss mechanism(5%) in 5 stocks picked for Task 2 and compare it with previous strategy.


```python
# Maximum drawdown is the largest percentage drop from the peak to the trough in the stock's price over a specific period.
```

```python
# Fetch historical data for all 50 stocks

nifty50_tickers = [
    "RELIANCE.NS", "TCS.NS", "HDFCBANK.NS", "INFY.NS", "ICICIBANK.NS", "HINDUNILVR.NS",
    "KOTAKBANK.NS", "ITC.NS", "LT.NS", "SBIN.NS", "AXISBANK.NS", "BAJFINANCE.NS",
    "BHARTIARTL.NS", "ASIANPAINT.NS", "HDFCBANK.NS", "ADANIGREEN.NS", "NTPC.NS",
    "ULTRACEMCO.NS", "ONGC.NS", "POWERGRID.NS", "JSWSTEEL.NS", "SUNPHARMA.NS",
    "HCLTECH.NS", "TITAN.NS", "MARUTI.NS", "INDUSINDBK.NS", "GRASIM.NS", "BAJAJ-AUTO.NS",
    "WIPRO.NS", "TATASTEEL.NS", "DIVISLAB.NS", "HEROMOTOCO.NS", "DRREDDY.NS", "M&M.NS",
    "NESTLEIND.NS", "COALINDIA.NS", "BRITANNIA.NS", "CIPLA.NS", "EICHERMOT.NS",
    "ADANIPORTS.NS", "BPCL.NS", "SHREECEM.NS", "HDFCLIFE.NS", "TECHM.NS", "TATAMOTORS.NS",
    "APOLLOHOSP.NS", "BAJAJFINSV.NS", "UPL.NS", "SBILIFE.NS"
]

data = yf.download(nifty50_tickers, start="2024-03-01", end="2024-09-01")['Adj Close']  # Time period is 6 months and fetching only 'Adj Close' as it shows effective closing price.

# Function to calculate maximum drawdown
def calculate_max_drawdown(stock_prices):
    cumulative_max = stock_prices.cummax()
    drawdown = (stock_prices - cumulative_max) / cumulative_max
    max_drawdown = drawdown.min()  # Most negative value
    return max_drawdown

# Calculate maximum drawdown for each stock
max_drawdowns = {}
for ticker in nifty50_tickers:
    max_drawdowns[ticker] = calculate_max_drawdown(data[ticker])

# Display results
max_drawdowns = pd.DataFrame(list(max_drawdowns.items()), columns=['Ticker', 'Max Drawdown'])
max_drawdowns
```

```python
# A stop-loss is a risk management strategy where a trade is exited when the stock drops by a specified percentage (5% in this case) from its purchase price. I'll compare this strategy against the simple buy-and-hold approach.
```

```python
from os import set_blocking

initial_price = []
sell =[]
current_price = []
final_price = []
buy_and_hold_return= []
stop_loss_return = []

for ticker in nifty50_tickers:

  # Define stop-loss threshold (5%)
  stop_loss_pct = 0.05

  # Simulate stop-loss strategy
  i_p = data[ticker].iloc[0]
  initial_price.append(i_p)

  holding = True  # Assume we start by holding the stock
  sell_price = None
  stop_loss_price = i_p * (1 - stop_loss_pct)

  # Iterate through the price data to check for stop-loss trigger
  for i in range(1, len(data[ticker])):
      c_p = data[ticker].iloc[i]
      current_price.append(c_p)

      # Check if stop-loss is triggered
      if holding and c_p < stop_loss_price:
          s_p = c_p
          sell.append(s_p)
          holding = False
          break  # Exit trade on stop-loss

  # If still holding, final price is the latest price
  f_p = s_p if s_p else data[ticker].iloc[-1]
  final_price.append(f_p)

  # Calculate returns
  b_h = (data[ticker].iloc[-1] - i_p) / i_p
  buy_and_hold_return.append(b_h*100)
  s_l = (f_p - i_p) / i_p
  stop_loss_return.append(s_l*100)



```

```python
d_ata = {'Ticker':nifty50_tickers, 'Buy and Hold Return':buy_and_hold_return, 'Stop Loss Return':stop_loss_return}

df = pd.DataFrame(d_ata)
df['Buy and Hold Return(%)'] = df['Buy and Hold Return'].round(2)
df['Stop Loss Return(%)'] = df['Stop Loss Return'].round(2)

df = df.drop(columns=['Buy and Hold Return', 'Stop Loss Return'])
```
